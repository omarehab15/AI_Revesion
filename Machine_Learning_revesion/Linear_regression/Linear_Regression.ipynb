# Univariate Linear regression 

```python
import numpy as np

X = data['RM'].values.reshape(-1, 1)
Y = data['MEDV'].values.reshape(-1,1)

def model(x , w , b):
    return w*x + b

def loss(w , b , x ,y):
    y_pred = model(x , w , b)
    return 0.5 * np.mean((y-y_pred)**2)

def gradient(w , b , x , y ) : 
    y_pred = model(x , w , b)
    dw = np.mean((y_pred - y) * x)
    db = np.mean(y_pred - y)
    return dw , db

def fit (x , y , w , b , learning_rate = 0.001 , epochs = 500000):
    error = []
    for i in range(epochs):
        dw , db = gradient(w , b ,x ,y)
        w -= learning_rate* dw
        b -= learning_rate* db
        error.append(loss(w,b,x ,y))
    return w , b , error


w = 0
b = 0
w , b , err  = fit(X , Y , w  , b )


    
```

```python
print("Coefficient (w):", w)
print("Intercept (b):", b)
# print error
print("Final loss:", err[-1]) 
```

# Multufeature Linear Regression

```python
# plot a heat map of the correlation matrix

plt.figure(figsize=(12,10)) 
corr = data.select_dtypes(include=[np.number]).corr()

corr_sorted = corr.sort_values('MEDV', ascending=False)

sns.heatmap(corr_sorted, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
def model(X, w, b):
    return X.dot(w) + b  # X: (n,m), w: (m,1), b: scalar

def loss(w, b, X, y):
    y_pred = model(X, w, b)
    return 0.5 * np.mean((y - y_pred) ** 2)

def gradient(w, b, X, y):
    m = len(y)
    y_pred = model(X, w, b)
    dw = (1/m) * X.T.dot(y_pred - y)  # dw: (m,1)
    db = np.mean(y_pred - y)
    return dw, db

def fit(X, y, w, b, learning_rate=0.01, epochs=1000):
    error = []
    for i in range(epochs):
        dw, db = gradient(w, b, X, y)
        w -= learning_rate * dw
        b -= learning_rate * db
        error.append(loss(w, b, X, y))
        if i % 100 == 0:
            print(f"Iteration {i}: Cost {error[-1]}")
    return w, b, error
```

```python
feature = ['RM' , 'LSTAT' , 'PTRATIO']
X = data[feature].values
Y = data['MEDV'].values.reshape(-1, 1)

# Feature scaling 

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

m,n = X_scaled.shape
w = np.zeros((n,1))
b = 0 

w_final, b_final, cost_history = fit(X_scaled, Y, w, b, learning_rate=0.01, epochs=1000)
print("Final weights (w):", w_final.ravel())
print("Final bias (b):", b_final)

y_pred = model(X_scaled, w_final, b_final)

```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y, y_pred)
rmse = np.sqrt(mse)
print(f"MSE: {mse:.3f}, RMSE: {rmse:.3f}")

```

```python
plt.figure(figsize=(10, 10))
plt.plot(range(len(cost_history)), cost_history)
plt.xlabel('epochs')
plt.ylabel('Cost')
plt.title('Cost Function Convergence - Multi-feature')
plt.show()
```
